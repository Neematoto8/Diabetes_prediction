```python
import pandas as pd # Imports pandas library with the alias 'pd'
data=pd.read_csv('/content/drive/MyDrive/DATASET/new_diabetes3_data_numerical.csv')
```

```python
data.size
```

```python
data.shape
```

```python
data.head(3)
```

```python

X=data.drop(['class'],axis=1)
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=HistGradientBoostingClassifier()
```

```python
model.fit(x_train,y_train)

accuracy=model.score(x_test,y_test)
accuracy
```

```python
y_pred=model.predict(x_test)
scores=f1_score(y_test,y_pred,average='weighted')
scores
```

```python
joblib.dump(model,'diabetes_model3')
print('model saved')
```
